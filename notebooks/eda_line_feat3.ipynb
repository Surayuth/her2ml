```python
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pl.read_csv("/home/surayuth/her2/extracted_features/linefeat_v3|scale_0.5|minhole_10|n_cell_100|n_cell_200|min_cont_0.05.csv") \
    .with_columns(
        (pl.col("h_max") / pl.col("n_cell_100")).alias("ratio_hole1"),
        (pl.col("h_max") / pl.col("n_cell_200")).alias("ratio_hole2")
    ) 
```

```python
df
```

```python
sns.scatterplot(data=df.filter(pl.col("ratio_hole2") >= 1), x="n_cell_200", y="h_max", s=5, hue="ihc_score")
```

```python
sns.scatterplot(data=df.filter(pl.col("ratio_hole1") >= 1), x="n_cell_100", y="h_max", s=5, hue="ihc_score")
```

```python
df \
    .filter(pl.col("ratio_hole2") >= 1) \
    .group_by("ihc_score") \
    .agg(
        pl.len().alias("count")
    ) \
    .sort("ihc_score")
```

```python
df \
    .filter(pl.col("ratio_hole2") >= 1) \
    .filter(pl.col("ihc_score") == "3+") \
    .group_by("case") \
    .agg(
        pl.len().alias("count"),
        pl.col("path").first()
    ) \
    .sort("case")
```

```python
df \
    .filter(
        (pl.col("ihc_score") == "1+") &
        (pl.col("ratio_hole2") >= 1)
    ) \
    .to_pandas()
```
