```python
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pl.read_csv("/home/surayuth/her2/extracted_features/linefeat_v3|scale_0.5|minhole_10|n_cell_10|n_cell_30|min_cont_0.1.csv") \
    .with_columns(
        (pl.col("h_max") / pl.col("n_cell_10")).alias("ratio_hole1"),
        (pl.col("h_max") / pl.col("n_cell_30")).alias("ratio_hole2")
    ) 
```

```python
df
```

```python
sns.scatterplot(data=df.filter(pl.col("ratio_hole1") >= 1), x="n_cell_10", y="h_max", s=5, hue="ihc_score")
```

```python
df \
    .filter(pl.col("ratio_hole2") >= 1) \
    .group_by("ihc_score") \
    .agg(
        pl.len().alias("count")
    ) \
    .sort("ihc_score")
```

```python
agg = df \
    .group_by("case") \
    .agg(
        pl.col("ihc_score").first(),
        pl.col("label").first(),
        pl.col("n").mean(),
        pl.col("h_max").mean(),
        pl.col("n_cell_10").mean(),
        pl.col("n_cell_30").mean(),
        pl.col("ratio_hole1").mean(),
        pl.col("ratio_hole2").mean()
    ) \
    .sort("label", "ihc_score", "case") 

agg
```

```python
sns.scatterplot(data=agg, x="ratio_hole1", y="n", hue="ihc_score", s=10)
plt.xlim([-0.1, 1])
```

```python
agg \
    .filter(pl.col("ihc_score") == "2+") \
    .filter(
        pl.col("ratio_hole1") < 0.1
    )
```

```python
agg3 = df \
    .filter(
        pl.col("ratio_hole1") > 0
    ) \
    .group_by("case") \
    .agg(
        pl.col("ihc_score").first(),
        pl.col("label").first(),
        pl.col("n").mean(),
        pl.col("h_max").mean(),
        pl.col("n_cell_10").mean(),
        pl.col("n_cell_30").mean(),
        pl.col("ratio_hole1").mean(),
        pl.col("ratio_hole2").mean()
    ) \
    .sort("label", "ihc_score", "case") 

agg3

sns.scatterplot(data=agg3, x="ratio_hole2", y="n", hue="ihc_score", s=10)
plt.xlim([-0.1, 1])
```

```python
th = 0.2

df \
    .group_by("case") \
    .agg(
        pl.when(pl.col("ratio_hole1") > th).then(pl.lit(1)).otherwise(pl.lit(0)).sum().alias(f"c>{th}"),
        pl.when(pl.col("ratio_hole1") == 0).then(pl.lit(1)).otherwise(pl.lit(0)).sum().alias("c=0"),
        pl.len().alias("count"),
        pl.col("ihc_score").first(),
        pl.col("label").first(),
        pl.col("n").mean(),
    ) \
    .filter(
        pl.col("ihc_score").is_in(["2-", "2+"])
    ) \
    .with_columns(
        (pl.col(f"c>{th}") / pl.col("count"))
        .alias(f"ratio_count>{th}"),
    ) \
    .with_columns(
        pl.lit("....")
        .alias("reason")
    ) \
    .sort(f"ratio_count>{th}") \
    .to_pandas() \
    .head(n=50)
```

```python
# focus on high contrast (n >= 5)

th = 0.2

df \
    .group_by("case") \
    .agg(
        pl.when(pl.col("ratio_hole1") > th).then(pl.lit(1)).otherwise(pl.lit(0)).sum().alias(f"c>{th}"),
        pl.when(pl.col("ratio_hole1") == 0).then(pl.lit(1)).otherwise(pl.lit(0)).sum().alias("c=0"),
        pl.len().alias("count"),
        pl.col("ihc_score").first(),
        pl.col("label").first(),
        pl.col("n").mean(),
    ) \
    .filter(
        pl.col("ihc_score").is_in(["2-", "2+"])
    ) \
    .with_columns(
        (pl.col(f"c>{th}") / pl.col("count"))
        .alias(f"ratio_count>{th}"),
    ) \
    .with_columns(
        pl.lit("....")
        .alias("reason")
    ) \
    .sort(f"ratio_count>{th}") \
    .filter(pl.col("n") >= 5) \
    .to_pandas() \
    .head(n=50)
```
