```python
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pl.read_csv("/home/surayuth/her2/extracted_features/linefeat_v3|scale_0.5|minhole_10.csv") \
    .with_columns(
        (pl.col("h_max") / pl.col("n_cell_100")).alias("ratio_hole1"),
        (pl.col("h_max") / pl.col("n_cell_200")).alias("ratio_hole2")
    ) 
```

```python
df \
    .with_columns(
        pl.when(pl.col("n") >= 1).then(pl.lit(1)).otherwise(pl.lit(0)).alias("n >= 1"),
        pl.when(pl.col("n") >= 2).then(pl.lit(1)).otherwise(pl.lit(0)).alias("n >= 2"),
        pl.when(pl.col("n") >= 3).then(pl.lit(1)).otherwise(pl.lit(0)).alias("n >= 3"),
        pl.when(pl.col("n") >= 4).then(pl.lit(1)).otherwise(pl.lit(0)).alias("n >= 4"),
        pl.when(pl.col("n") >= 5).then(pl.lit(1)).otherwise(pl.lit(0)).alias("n >= 5"),
        pl.when(pl.col("n") >= 6).then(pl.lit(1)).otherwise(pl.lit(0)).alias("n >= 6"),
        pl.when(pl.col("n") >= 7).then(pl.lit(1)).otherwise(pl.lit(0)).alias("n >= 7"),
        pl.when(pl.col("n") >= 8).then(pl.lit(1)).otherwise(pl.lit(0)).alias("n >= 8"),
        pl.when(pl.col("n") >= 9).then(pl.lit(1)).otherwise(pl.lit(0)).alias("n >= 9"),
        pl.when(pl.col("n") >= 10).then(pl.lit(1)).otherwise(pl.lit(0)).alias("n >= 10")
    ) \
    .group_by("ihc_score") \
    .agg(
        pl.col("n >= 1").sum(),
        pl.col("n >= 2").sum(),
        pl.col("n >= 3").sum(),
        pl.col("n >= 4").sum(),
        pl.col("n >= 5").sum(),
        pl.col("n >= 6").sum(),
        pl.col("n >= 7").sum(),
        pl.col("n >= 8").sum(),
        pl.col("n >= 9").sum(),
        pl.col("n >= 10").sum(),
    ) \
    .sort("ihc_score")
```
